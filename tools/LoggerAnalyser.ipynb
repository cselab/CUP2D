```python
# Compute non-zero indices and classify the edge type they belong to
def logDiffDf(cpuLog, gpuLog, key = "Ax0"):
    
    dfs = [None] * len(gpuLog.index)
    
    for rank, row in gpuLog.iterrows():

        cpuAx = np.asarray(cpuLog[key][rank], dtype=np.double)
        gpuAx = np.asarray(gpuLog[key][rank], dtype=np.double)
    
        nonzero_idx = np.where(np.logical_not(np.isclose(cpuAx, gpuAx, atol=1e-10)))[0]
        block_idx = np.floor_divide(nonzero_idx, 64)
        cellerr_idx = np.mod(nonzero_idx, 64)
        err_ix = np.mod(cellerr_idx, 8)
        err_iy = np.floor_divide(cellerr_idx,8)
        has_bd_elems = np.in1d(nonzero_idx, gpuLog["bdCooRowA"][rank])
        
        dfs[rank] = pd.DataFrame(
                    {key+"CPU": cpuAx[nonzero_idx],
                     key+"GPU": gpuAx[nonzero_idx],
                     "Error": (cpuAx[nonzero_idx] - gpuAx[nonzero_idx]).flatten(),
                     "CellId": nonzero_idx.flatten(),
                     "BlockId": block_idx.flatten(), 
                     "ix": err_ix.flatten(), 
                     "iy": err_iy.flatten(),
                     "bdRow" : has_bd_elems})
    dfExt = None
    
    if len(gpuLog.index) == 1:
        cooValA = np.asarray(gpuLog["locCooValA"][0], dtype=np.double)
        cooRowA = np.asarray(gpuLog["locCooRowA"][0], dtype=np.int32)
        cooColA = np.asarray(gpuLog["locCooColA"][0], dtype=np.int32)
        l = []
        for index, row in dfs[0].iterrows():
            cell_glob_idx = row["CellId"]
            cell_row_idx = cooRowA == cell_glob_idx
            col_glob_idx = np.where(cell_row_idx)[0]
            col_glob = cooColA[col_glob_idx]
            block_idx = np.floor_divide(col_glob, 64)
            col_loc = np.mod(col_glob, 64)
            col_err_ix = np.mod(col_loc, 8)
            col_err_iy = np.floor_divide(col_loc,8)
            valA = cooValA[cell_row_idx]
            colA = cooColA[cell_row_idx]
            l.append(
                {"val": valA.flatten().round(3), 
                 "col": colA.flatten(), 
                 "col_CellID": col_glob.flatten(), 
                 "col_BlockId": block_idx.flatten(),
                 "col_ix": col_err_ix.flatten(),
                 "col_iy": col_err_iy.flatten()})
        dfs.append(pd.DataFrame(l))
    

    return dfs

def logDiffDf2xlsx(cpuLog, gpuLog, key): 
    diffDf = logDiffDf(cpuLog, gpuLog, key)

    # Create a Pandas Excel writer using XlsxWriter as the engine.
    writer = pd.ExcelWriter("diff_df_" + key + ".xlsx", engine='xlsxwriter')
    # Write each dataframe to a different worksheet.
    for r in range(len(diffDf)):
        diffDf[r].to_excel(writer, sheet_name="Rank" + str(r))
    writer.save()
```
